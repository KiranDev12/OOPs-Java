# Interfaces

    Interfaces are keywors used in java to fully isolate the interface from its implementation. It provides a set of abstract methods and constants which can be implemented by as many as class in their own way. This completes "one interface, multiple methods".

* It is used to achieve fully abstraction.
* By interface, we can support the functionality of multiple inheritance.
* It can be used to achieve loose coupling.

public abstract is added before the methods and <br>
public static final before data members.
